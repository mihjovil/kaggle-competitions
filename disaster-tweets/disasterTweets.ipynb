```python
from sklearn.model_selection import train_test_split
import keras
```

```python
train_df, val_df, train_y, val_y = train_test_split(train_df.text, train_df.target, test_size=0.2)
```

```python
tokenizer = keras.preprocessing.text.Tokenizer(num_words=20, lower=True)
```

```python
fitted_train = tokenizer.fit_on_texts(train_df)
```

```python
sequences_train = tokenizer.texts_to_sequences(train_df)
sequences_test = tokenizer.texts_to_sequences(val_df)
```

```python
max_length = 5
sequences_train = keras.preprocessing.sequence.pad_sequences(sequences_train, maxlen=max_length, truncating='pre', padding='pre',value=-1)
sequences_test = keras.preprocessing.sequence.pad_sequences(sequences_test, maxlen=max_length, truncating='pre', padding='pre',value=-1)
```

```python
sequences_train.shape, sequences_test.shape, train_y.shape, val_y.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(sequences_train, train_y)
```

```python
lr.score(sequences_train, train_y)
```

```python
lr.score(sequences_test, val_y)
```
